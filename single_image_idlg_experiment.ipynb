```python
def tensor_to_np(tensor):
    if tensor.dim() == 4 and tensor.shape[0] == 1:
        tensor = tensor.squeeze(0).permute(1, 2, 0)
    return tensor.detach().cpu().numpy()

iterations_to_plot = [5, 10, 15, 20, 25, 30]
num_rows = 2
num_iterations = len(iterations_to_plot)
num_images_to_plot = num_iterations + 3

fig, axes = plt.subplots(num_rows, num_images_to_plot, figsize=(20, num_rows * 4))

row_titles = ['No def', 'Ours']

for i in range(num_rows):
    axes[i, 0].text(0.5, 0.5, row_titles[i], fontsize=20, ha='center', va='center', transform=axes[i, 0].transAxes)
    axes[i, 0].axis('off')

axes[0, 1].imshow(tensor_to_np(dummy_data_baseline))
axes[0, 1].set_title("Dummy Data")
axes[0, 1].axis('off')

axes[0, 2].imshow(prepare_tensor_for_plotting(gt_data_orig_baseline))
axes[0, 2].set_title("Ground Truth")
axes[0, 2].axis('off')

for j, idx in enumerate(iterations_to_plot):
    recon_image = history_baseline[idx - 1]
    col_index = j + 3 
    axes[0, col_index].imshow(recon_image)
    iteration_num = idx * 10 
    axes[0, col_index].set_title(f"Iter={iteration_num}")
    axes[0, col_index].axis('off')

axes[1, 1].imshow(tensor_to_np(dummy_data_0_1_0))
axes[1, 1].set_title("Dummy Data")
axes[1, 1].axis('off')

axes[1, 2].imshow(prepare_tensor_for_plotting(gt_data_orig_0_1_0))
axes[1, 2].set_title("Ground Truth")
axes[1, 2].axis('off')

for j, idx in enumerate(iterations_to_plot):
    recon_image = history_0_1_0[idx - 1]
    col_index = j + 3
    axes[1, col_index].imshow(recon_image)
    iteration_num = idx * 10
    axes[1, col_index].set_title(f"Iter={iteration_num}")
    axes[1, col_index].axis('off')

#plt.tight_layout()
plt.show()
```
